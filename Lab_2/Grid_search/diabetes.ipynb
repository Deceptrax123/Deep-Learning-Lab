```python
from keras.models import Sequential 
from keras.layers import Dense
from sklearn.model_selection import StratifiedKFold
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np
import pandas as pd 
import keras
```

```python
dataset=pd.read_csv("/Volumes/T7 Shield/Smudge/Datasets/diabates/diabetes.csv")
```

```python
#X and Y
X=np.array(dataset.drop('Outcome',axis=1))
Y=np.array(dataset['Outcome'])
```

```python
def create_model():
    model=Sequential()
    model.add(Dense(12,input_dim=8,kernel_initializer='uniform',activation='relu'))
    model.add(Dense(1,kernel_initializer='uniform',activation='sigmoid'))

    model.compile(loss=keras.losses.BinaryCrossentropy(),metrics=keras.metrics.BinaryAccuracy(),optimizer='adam')

    return model
```

```python
model=KerasClassifier(build_fn=create_model,verbose=0)

optimizer=['rmsprop','adam']
epochs=[2,5,10]
batch=[8,16,32]

grid_params=dict(optimizer=optimizer,epochs=epochs,batch_size=batch)
gridsearch=GridSearchCV(estimator=model,param_grid=grid_params)

result=gridsearch.fit(X,Y,verbose=0)
```

```python
print("Best scores")
print(result.best_score_,result.best_params_)

mean=result.cv_results_['mean_test_score']
stds=result.cv_results_['std_test_score']

params=result.cv_results_['params']
for mean,std,params in zip(mean,stds,params):
    print("%f (%f) with :%r" %(mean,std,params))
```
