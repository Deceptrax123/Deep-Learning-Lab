```python
from keras.models import Sequential 
from keras.layers import Dense
from sklearn.model_selection import StratifiedKFold
import numpy as np
import pandas as pd 
import keras
```

```python
dataset=pd.read_csv("/Volumes/T7 Shield/Smudge/Datasets/diabates/diabetes.csv")
```

```python
dataset
```

```python
#X and Y
X=np.array(dataset.drop('Outcome',axis=1))
Y=np.array(dataset['Outcome'])
```

```python
kfold=StratifiedKFold(n_splits=10,shuffle=True,random_state=7)
tot_scores=[]

for train,test in kfold.split(X,Y):
    model=Sequential()
    model.add(Dense(12,input_dim=8,kernel_initializer='uniform',activation='relu'))
    model.add(Dense(1,kernel_initializer='uniform',activation='sigmoid'))

    model.compile(loss=keras.losses.BinaryCrossentropy(),metrics=keras.metrics.BinaryAccuracy(),optimizer='adam')

    model.fit(X[train],Y[train],epochs=100,batch_size=10,verbose=0)

    scores=model.evaluate(X[test],Y[test],verbose=0)
    tot_scores.append(scores[1]*100)
    
```

```python
sum(tot_scores)/len(tot_scores)
```
