```python
import tensorflow as tf
import keras
from keras.models import Sequential
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt 
```

```python
#Make the circular dataset
X,y=make_circles(n_samples=1000,factor=0.5,noise=0.1)
plt.figure(figsize=(8,6))
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
#Neural Network
nn=Sequential()
nn.add(tf.keras.layers.Dense(units=5,activation='relu',input_shape=(2,)))
nn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
nn=Sequential()
nn.add(tf.keras.layers.Dense(units=5,activation='sigmoid',input_shape=(2,)))
nn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
nn=Sequential()
nn.add(tf.keras.layers.Dense(units=5,activation='sigmoid',input_shape=(2,)))
nn.add(tf.keras.layers.Dense(units=10,activation='sigmoid'))
nn.add(tf.keras.layers.Dense(units=20,activation='sigmoid'))
nn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
nn.compile(optimizer='adam',loss=keras.losses.BinaryCrossentropy(),metrics=keras.metrics.BinaryAccuracy())
```

```python
nn.fit(X,y,batch_size=16,epochs=50,verbose=True)
```
